*  The Open University is one of the largest distance learning instituations,
serving diverse learners across age groups, employment backgrounds, and study intensities.<br> As online education scales, leadership focus is shifting from measuring simple activity (logins, clicks) to measuring Acdemic Success Quality - ensuring that engagment with learning materials translates into meaningful academic outcomes and sustained retention.

##### **Objectives:** <br>
*  Analyze student learning data from the Open University to evaluate **Academic Success Efficiency (ASE)** and key drivers of student performance and withdrawal.
*  Examine how ASE varies across **demographic, socioeconomic, behavioural and course related factors.**
*  Assess engagement patterns, assessment preparation intensity, and VLE activity interactions across modules.
*  Identify **high efficient learners, inefficient learners, and students at risk of withdrawal.**
*  Generate actionable insights that support early intervention and improve student success outcomes.

### **Finding missing values**

```python
def missing_values(df, name):
  missing = df.isnull().sum()
  missing_percentage = (missing / len(df)) * 100
  missing_df = pd.DataFrame({'Missing Count': missing, 'Missing %': missing_percentage.round(2)})

  print(f"Missing values in {name}:")
  print(missing_df[missing_df['Missing Count']>0])

  print("-"*80)

missing_values(studentInfo, 'studentInfo')
missing_values(studentRegistration, 'studentRegistration')
missing_values(courses, 'courses')
missing_values(assessments, 'assessments')
missing_values(studentAssessment, 'studentAssessment')
missing_values(studentVle, 'studentVle')
missing_values(vle, 'vle')
```

### **MERGING DATASETS - PREPARING FINAL DATASET**

```python
# 1.Preparing assessment data with types and weights
# merging 'studentAssessment' with 'assessments'
assessment_full = pd.merge(
    studentAssessment,
    assessments[['id_assessment', 'code_module', 'code_presentation', 'assessment_type', 'weight']], on='id_assessment', how='left')


# 2.Calculate weighted scores
# calculating weighted score for each assessment
assessment_full['weighted_score'] = assessment_full['score'] * (assessment_full['weight'] / 100)

# aggregate by student
student_assessment_agg = assessment_full.groupby(
    ['id_student', 'code_module', 'code_presentation']).agg({
    'score': 'mean',
    'weighted_score' : 'sum',
    'weight' : 'sum',
    'id_assessment' : 'count'
}).reset_index()

student_assessment_agg.columns = [
    'id_student', 'code_module', 'code_presentation','avg_score', 'weighted_score_sum','total_weight', 'submission_count']

# calculating weighted average score
student_assessment_agg['weighted_avg_score'] = np.where(
    student_assessment_agg['total_weight'] > 0, (student_assessment_agg['weighted_score_sum'] / student_assessment_agg['total_weight']) * 100, 0)


# 3.Calculating scores by assessment type
score_by_type = assessment_full.groupby(
    ['id_student', 'code_module', 'code_presentation', 'assessment_type'])['score'].mean().reset_index()

score_pivot = score_by_type.pivot_table(
    index=['id_student', 'code_module', 'code_presentation'],
    columns = 'assessment_type',
    values = 'score',
    aggfunc = 'mean'
).reset_index()

score_pivot.columns.name = None

# Ensuring all assessment tye columns exist
for col in ['TMA', 'CMA', 'Exam']:
  if col not in score_pivot.columns:
    score_pivot[col] = 0
  else:
    score_pivot[col] = score_pivot[col].fillna(0)


# 4.Merge assessment metrics
assessment_metrics = pd.merge(student_assessment_agg, score_pivot, on=['id_student', 'code_module', 'code_presentation'], how='outer')

assessment_metrics =  assessment_metrics.fillna(0)

# calculating VLE engagement metrics
vle_metrics = studentVle.groupby(
    ['id_student', 'code_module', 'code_presentation']
).agg({
    'sum_click': 'sum',
    'date': 'nunique'
}).reset_index()

vle_metrics.columns = ['id_student', 'code_module', 'code_presentation', 'total_clicks', 'active_days']


# 5.Creating complete student dataset
student_complete = studentInfo.copy()

# adding registration info
student_complete = pd.merge(
    student_complete,
    studentRegistration,
    on=['id_student', 'code_module', 'code_presentation'],
    how='left')

# adding course info
student_complete = pd.merge(
    student_complete,
    courses,
    on=['code_module', 'code_presentation'],
    how='left')

# adding VLE metrics
student_complete = pd.merge(
    student_complete,
    vle_metrics,
    on=['id_student', 'code_module', 'code_presentation'],
    how='left')

# adding assessment metrics
student_complete = pd.merge(
    student_complete,
    assessment_metrics,
    on=['id_student', 'code_module', 'code_presentation'],
    how='left')


# handling nulls
# VLE metrics: NULL = No engagement = 0
student_complete['total_clicks'] = student_complete['total_clicks'].fillna(0)
student_complete['active_days'] = student_complete['active_days'].fillna(0)

# Assessment metrics: NULL = No submission = 0
student_complete['avg_score'] = student_complete['avg_score'].fillna(0)
student_complete['weighted_avg_score'] = student_complete['weighted_avg_score'].fillna(0)
student_complete['submission_count'] = student_complete['submission_count'].fillna(0)
student_complete['TMA'] = student_complete['TMA'].fillna(0)
student_complete['CMA'] = student_complete['CMA'].fillna(0)
student_complete['Exam'] = student_complete['Exam'].fillna(0)

# Dropping missing values rows for 'date_registration', 'date_unregistration'
student_complete = student_complete.dropna(subset=['date_registration'])

# Dropping 'weighted_score_sum', 'total_weight' temporary columns
student_complete = student_complete.drop(columns=['weighted_score_sum', 'total_weight'], errors='ignore')

print("Dataset is ready")

# displaying dataset
pd.set_option('display.max_columns', None)
student_complete.head()
```

### **KEY METRICS CALCULATION**

#### **Academic Performance (AP)**

```python
# Normalizing weighted average score
min_score = student_complete['weighted_avg_score'].min()
max_score = student_complete['weighted_avg_score'].max()

if max_score > min_score:
  student_complete['score_normalized'] = (
      (student_complete['weighted_avg_score'] - min_score) / (max_score - min_score)
  )
else:
  student_complete['score_normalized'] = 0

# creating completion indicator
def get_completion_indicator(result):
  if result in ['Pass', 'Distinction']:
    return 1.0
  elif result == 'Fail':
    return 0.5
  else:
    return 0.0

student_complete['completion_indicator'] = student_complete['final_result'] .apply(get_completion_indicator)


# pass/fail status per assessmnet -> passed assesment ratio
# (number of assessmnets passed / total submitted, per student-module)
assessment_full['is_passed'] = (assessment_full['score']>=40).astype(int)
passed_ratio = assessment_full.groupby(['id_student', 'code_module', 'code_presentation']
).agg(
    passed_count=('is_passed', 'sum'),
    total_count=('is_passed', 'count')
).reset_index()

passed_ratio['passed_assessment_ratio'] = (passed_ratio['passed_count']/passed_ratio['total_count']).fillna(0)

student_complete = student_complete.drop(columns=['passed_assessment_ratio'], errors='ignore')
student_complete = pd.merge(
    student_complete,
    passed_ratio[['id_student', 'code_module', 'code_presentation', 'passed_assessment_ratio']],
    on=['id_student', 'code_module', 'code_presentation'],
    how='left'
)
student_complete['passed_assessment_ratio'] = student_complete['passed_assessment_ratio'].fillna(0)

# weighted score across assessment types
student_complete['type_weighted_score'] = (
    student_complete['Exam'] * 0.50 +
    student_complete['TMA'] * 0.30 +
    student_complete['CMA'] * 0.20
) / 100


# Calculating 'Academic Performance (AP)'
student_complete['academic_performance'] = (student_complete['score_normalized'] * 0.40 + student_complete['completion_indicator'] * 0.25 + student_complete['passed_assessment_ratio'] * 0.20 + student_complete['type_weighted_score'] * 0.15)
print("Academic Performance (AP) Statistics:")
print(student_complete['academic_performance'].describe().round(3))
```

#### **Learning Effort (LE)**

```python
# normalizing clicks (0-1)
min_clicks = student_complete['total_clicks'].min()
max_clicks = student_complete['total_clicks'].max()

if max_clicks > min_clicks:
  student_complete['clicks_normalized'] = (
      (student_complete['total_clicks'] - min_clicks) / (max_clicks - min_clicks)
  )
else:
  student_complete['clicks_normalized'] = 0

# normalizing active days (0-1)
min_days = student_complete['active_days'].min()
max_days = student_complete['active_days'].max()

if max_days > min_days:
  student_complete['days_normalized'] = (
      (student_complete['active_days'] - min_days) / (max_days - min_days)
  )
else:
      student_complete['days_normalized'] = 0

# normalizing submissions (0-1)
min_subs = student_complete['submission_count'].min()
max_subs = student_complete['submission_count'].max()

if max_subs > min_subs:
  student_complete['submissions_normalized'] = (
      (student_complete['submission_count'] - min_subs) / (max_subs - min_subs)
  )
else:
  student_complete['submissions_normalized'] = 0


# registration duration -> time from registration to end of module
student_complete['registration_duration'] = (
    student_complete['module_presentation_length'] - student_complete['date_registration']).clip(lower=0)

min_reg = student_complete['registration_duration'].min()
max_reg = student_complete['registration_duration'].max()

student_complete['registration_duration_normalized'] = (
    (student_complete['registration_duration'] - min_reg) / (max_reg - min_reg)
    if max_reg > min_reg else 0
)


# number of enrolled modules per student
modules_per_student = (
    student_complete.groupby(['id_student'])['code_module'].nunique().reset_index().rename(columns={'code_module' : 'num_enrolled_modules'})
)

student_complete = student_complete.drop(columns=['num_enrolled_modules'], errors='ignore')
student_complete = pd.merge(student_complete, modules_per_student, on='id_student', how='left')

min_mod = student_complete['num_enrolled_modules'].min()
max_mod = student_complete['num_enrolled_modules'].max()

student_complete['modules_normalized'] = (
    (student_complete['num_enrolled_modules'] - min_mod) / (max_mod - min_mod)
    if max_mod > min_mod else 0
)


# calculating 'Learning effort (LE)'
student_complete['learning_effort'] = (
    student_complete['clicks_normalized'] * 0.30 +
    student_complete['days_normalized'] * 0.25 +
    student_complete['submissions_normalized'] * 0.20 +
    student_complete['registration_duration_normalized'] * 0.15 +
    student_complete['modules_normalized'] * 0.10
 )
print("Learning Effort (LE) Statistics:")
print(student_complete['learning_effort'].describe().round(3))
```

#### **Academic Success Efficiency (ASE)**

```python
# ASE = Academic Performance/Learning effort
epsilon = 0.05 # to avoid divison by zero error

student_complete['ASE'] = (
    student_complete['academic_performance']/(student_complete['learning_effort'] + epsilon)
)

print("ASE calculated \n")
# statistics
print("Statistics:")
print(student_complete['ASE'].describe().round(3))
```

### **Creating student segments**

```python
ase_low = student_complete['ASE'].quantile(0.25)
ase_high = student_complete['ASE'].quantile(0.75)

effort_median = student_complete['learning_effort'].median()

def categorize_student(row):
  ase = row['ASE']
  effort = row['learning_effort']

  if ase >= ase_high:
    return 'Efficient Learners'
  elif ase <= ase_low and effort >= effort_median:
    return 'Struggling Students - Hard workers'
  elif ase <= ase_low and effort < effort_median:
    return 'Disengaged/At-Risk'
  else:
    return 'Coginitive Overload Risk'

student_complete['student_segment'] = student_complete.apply(categorize_student, axis=1)
segment_percent = (student_complete['student_segment'].value_counts(normalize=True).mul(100).round(2).reset_index())
segment_percent.columns = ['student_segment', '(%)Total_Students']
segment_summary = (student_complete.groupby('student_segment')[['ASE', 'learning_effort']].median().round(3).reset_index())

segment_final = segment_summary.merge(segment_percent, on='student_segment')
print("Student Efficiency Segment Summary:\n")
segment_final
```

#### **1. Key Findings:**
*  **Cognitive Overload Risk (50% of students) -**
   This is the largest segment. These students show **high learning effort but only moderate academic efficiency,** indicating that student engagement with learning materials does not proportionally convert into academic outcomes.

*  **Efficient Learners (25% of students) -**
   This group achieves **high academic efficiency with relatively low effort,** suggesting effective learning strategies and strong understanding of course material.
*  **Disengaged / At-Risk Students (22.54 of students) -**
   These students demonstrate **very low learning effort and negligible academic efficiency,** indicating minimal interaction with course content and a high likelihood of academic failure or withdrawal.
*  **Struggling but Hard-Working Students (2.47% of students) -**
   Despite **high effort levels,** this group shows low academic efficiency, suggesting difficulties in understanding course material or ineffective study approaches.


#### **2. Business Insights:**
*  **Learning efficiency varies significantly across students. While 25% of learners are efficient**, a large proportion either requires excessive effort for moderate outcomes or shows weak academic performance.

*  **A notable segment of students is at academic risk.** Approximately 25% of learners (disengaged and struggling segments) show patterns associated with **assessment failure, withdrawal risk, or repeated module attempts.**
*  **High engagement does not always convert into strong outcomes. The large coginitive overload segment (50%)** indicates that increases interaction with learning materisals may not consistently lead to proportional academic success, suggesting that students may be not learning in the most effective way.













### **SUPPORTING METRICS**

##### **1. Assessment Effective Metrics**

```python
# 1.1 Pass Rate
assessment_full['is_passed'] = (assessment_full['score'] >= 40).astype(int)
pass_rate = assessment_full.groupby(
    ['id_student', 'code_module', 'code_presentation']
).agg(
    passes_count=('is_passed', 'mean'),
    attempted_count=('id_assessment', 'count')
).reset_index()
pass_rate.columns = ['id_student', 'code_module', 'code_presentation', 'passes_count', 'attempted_count']
pass_rate['pass_rate'] = (pass_rate['passes_count']/pass_rate['attempted_count']).round(3)

print(f"Pass Rate: :{pass_rate['pass_rate'].mean():.2%}\n")

# 1.2 Average Score by Type
print(f"TMA Avg: {student_complete['TMA'].mean():.2f}")
print(f"CMA Avg: {student_complete['CMA'].mean():.2f}")
print(f"Exam Avg: {student_complete['Exam'].mean():.2f}\n")

# 1.3 Submission Rate
total_assigned = assessments.groupby(
    ['code_module', 'code_presentation']
)['id_assessment'].nunique().reset_index()
total_assigned.columns = ['code_module', 'code_presentation', 'total_assigned']
submission_rate = pd.merge(
    pass_rate[['id_student', 'code_module', 'code_presentation', 'attempted_count']],
    total_assigned,
    on=['code_module', 'code_presentation'], how='left'
)
submission_rate['submission_rate'] = (submission_rate['attempted_count']/submission_rate['total_assigned']).clip(upper=1.0).round(3)

print(f"Submission Rate: {submission_rate['submission_rate'].mean():.2%}\n")

# 1.4 Performance Trend = Score progression across attempts
def calc_trend(scores):
  if len(scores) < 2:
    return 0
  return np.polyfit(range(len(scores)), scores, 1)[0]

assessment_sorted = assessment_full.sort_values(by=['id_student', 'code_module', 'date_submitted'])
performance_trend = assessment_sorted.groupby(
    ['id_student', 'code_module', 'code_presentation']
)['score'].apply(calc_trend).reset_index()
performance_trend.columns = ['id_student', 'code_module', 'code_presentation', 'performance_trend']
performance_trend['trend_category'] = pd.cut(
    performance_trend['performance_trend'],
    bins=[-np.inf, -2, 2, np.inf],
    labels=['Declining', 'Stable', 'Improving'],
)
print("Performance Trend: Scores over time")
percent = performance_trend['trend_category'].value_counts(normalize=True)*100
print(percent.round(1).astype(str)+'%')
```

##### **2. Engagement Health Metrics**

```python
# 2.1 VLE Intensity Index
student_complete['vle_intensity'] = np.where(
    student_complete['active_days'] > 0,
    student_complete['total_clicks'] / student_complete['active_days'],0).round(2)
print(f"VLE Intensity Index: {student_complete['vle_intensity'].mean():.2f}")

# 2.2 Early Engagement Score = Clicks in first 30 days
early_clicks = studentVle[studentVle['date'] <= 30].groupby(
    ['id_student', 'code_module', 'code_presentation'])['sum_click'].sum().reset_index()

early_clicks = early_clicks.rename(columns={'sum_click': 'early_clicks'})
student_complete = student_complete.drop(columns=['early_clicks'], errors='ignore')

student_complete = student_complete.merge(
    early_clicks,
    on=['id_student', 'code_module', 'code_presentation'],
    how='left')
student_complete['early_clicks'] = student_complete['early_clicks'].fillna(0)

student_complete['early_engagement_score'] = np.where(
    student_complete['active_days'] > 0,
    student_complete['early_clicks']/student_complete['total_clicks'], 0).round(3)

print(f"Early Engagement Score: {student_complete['early_engagement_score'].mean():.2f}")

# 2.3 Content Coverage Rate
total_activities = vle.groupby(
    ['code_module', 'code_presentation'])['id_site'].nunique().reset_index()
total_activities.columns = ['code_module', 'code_presentation', 'total_activities']

accessed_activities = studentVle.groupby(
    ['id_student', 'code_module', 'code_presentation'])['id_site'].nunique().reset_index()
accessed_activities.columns = ['id_student', 'code_module', 'code_presentation', 'accessed_activities']

content_coverage = pd.merge(accessed_activities, total_activities, on = ['code_module', 'code_presentation'], how='left')
content_coverage['content_coverage_rate'] = (
    content_coverage['accessed_activities']/content_coverage['total_activities']
).clip(upper=1.0).round(3)

print(f"Content Coverage Rate: {content_coverage['content_coverage_rate'].mean():.2%}")

# 2.4 Assessment Preparation Ratio
first_assessment = assessment_full.groupby(
    ['id_student', 'code_module', 'code_presentation'])['date_submitted'].min().reset_index()
first_assessment.columns = ['id_student', 'code_module', 'code_presentation', 'first_assessment_date']

vle_with_date = pd.merge(studentVle, first_assessment, on=['id_student', 'code_module', 'code_presentation'], how='left')
clicks_before = vle_with_date[vle_with_date['date'] < vle_with_date['first_assessment_date']].groupby(
      ['id_student', 'code_module', 'code_presentation']
)['sum_click'].sum().reset_index()

clicks_before.columns = ['id_student', 'code_module', 'code_presentation', 'clicks_before_assessment']

student_complete = student_complete.drop(columns=['clicks_before_assessment'], errors='ignore')
student_complete = student_complete.merge(
    clicks_before,
    on=['id_student', 'code_module', 'code_presentation'],
    how='left')

clicks_before['clicks_before_assessment'] = clicks_before['clicks_before_assessment'].fillna(0)
student_complete['assessment_preparation_ratio'] = np.where(
    student_complete['total_clicks'] > 0,
    student_complete['clicks_before_assessment']/student_complete['total_clicks'], 0).round(3)

print(f"Assessment Preparation Ratio: {student_complete['assessment_preparation_ratio'].mean():.2%}")
```

##### **3. Retention Risk Indicators**

```python
# 3.1 Withdrawal rate by module
withdrawal_by_module = (
                  studentInfo
                  .assign(is_withdrawn = studentInfo['final_result'].eq('Withdrawn'))
                  .groupby('code_module')['is_withdrawn'].mean().mul(100).round(2).reset_index(name='withdrawal_rate'))
print(f"Withdrawal Rates:\n {withdrawal_by_module.sort_values('withdrawal_rate', ascending=False)}")

# 3.2 Low Early Engagement + Low Submission Rate
# merging submission rate
student_complete = student_complete.drop(columns=['submission_rate'], errors='ignore')

student_complete = pd.merge(student_complete,
          submission_rate[['id_student', 'code_module', 'code_presentation', 'submission_rate']],
          on=['id_student', 'code_module', 'code_presentation'], how='left')
student_complete['submission_rate'] = student_complete['submission_rate'].fillna(0)

# calculating thresholds
low_engage_threshold = student_complete['early_engagement_score'].quantile(0.25)
low_submit_threshold = submission_rate['submission_rate'].quantile(0.25)

# creating flag
student_complete['flag_disengaged'] = (
    (student_complete['early_engagement_score'] < low_engage_threshold) &
    (student_complete['submission_rate'] < low_submit_threshold).astype(int)
)
print("\nLow Early engagement and Low Submission rate:")
print(f"Early Engagement < {low_engage_threshold} , Submission Rate < {low_submit_threshold}")
print(f"Disengaged Students: {student_complete['flag_disengaged'].sum()}")

# 3.3 High effort but declining scores
def calc_trend(scores):
  if len(scores) < 2:
    return 0
  return np.polyfit(range(len(scores)), scores, 1)[0]

assessment_sorted = assessment_full.sort_values(
    ['id_student', 'code_module', 'code_presentation', 'date_submitted']
)

performance_trend = assessment_sorted.groupby(['id_student', 'code_module', 'code_presentation'])['score'].apply(calc_trend).reset_index()
performance_trend.columns = ['id_student', 'code_module', 'code_presentation', 'performance_trend']

if 'performance_trend' in student_complete.columns:
  student_complete = student_complete.drop(columns=['performance_trend'])

# merging
student_complete = pd.merge(student_complete,
        performance_trend,
        on=['id_student', 'code_module', 'code_presentation'], how='left')
student_complete['performance_trend'] = student_complete['performance_trend'].fillna(0)

high_effort_th = student_complete['vle_intensity'].quantile(0.75)
declining_th = student_complete['performance_trend'].quantile(0.25)

student_complete['flag_struggling'] = (
    (student_complete['vle_intensity'] >= high_effort_th) &
    (student_complete['performance_trend'] < declining_th)).astype(int)

print("\nStruggling (High Effort + Declining Scores):")
print(f"Thresholds: VLE Intensity >= {high_effort_th:.2f}, Trend < {declining_th:.2f}")
print(f"Struggling students: {student_complete['flag_struggling'].sum()}")

# 3.4 Late Registration + Low VLE usage
late_reg_threshold = 0
low_vle_threshold = student_complete['vle_intensity'].quantile(0.25)
student_complete['flag_late_starter'] = (
    (student_complete['date_registration'] > late_reg_threshold) &
    (student_complete['total_clicks'] < low_vle_threshold)
).astype(int)

print("\nLate Registration and Low VLE:")
print(f"Thresholds: Registration > {late_reg_threshold}, Clicks < {low_vle_threshold}")
print(f"Late Starter Students: {student_complete['flag_late_starter'].sum()}\n")

print("Supporting Metrics Calculated")
```

#### **Segments by RISK LEVEL**

```python
# Risk Level calculation
student_complete['risk_score'] = (
    student_complete['flag_disengaged'] +
    student_complete['flag_struggling'] +
    student_complete['flag_late_starter']
)

# creating risk level categories
def assign_risk_level(score):
  if score >=2:
    return 'Critical'
  elif score == 1:
    return 'High Risk'
  else:
    return 'Low Risk'

student_complete['risk_level']  = student_complete['risk_score'].apply(assign_risk_level)

# calculate distribution
risk_distribution = student_complete['risk_level'].value_counts()
risk_percentage = (student_complete['risk_level'].value_counts(normalize=True)*100).round(2)

print("Risk level distribution:")
# printing risk distribution
for level in ['Low Risk', 'High Risk', 'Critical']:
  if level in risk_distribution.index:
    count = risk_distribution[level]
    percentage = risk_percentage[level]
    print(f"{level}: {count} students ({percentage}%)")

total_at_risk = (student_complete['risk_level'] != 'Low Risk').sum()
total_at_risk_percentage = (total_at_risk/len(student_complete)*100).round(2)

print(f"\nTotal at risk students: {total_at_risk} ({total_at_risk_percentage}%)")
print(f"Safe Students: {len(student_complete) - total_at_risk} ({100 - total_at_risk_percentage}%)")
```

### **HYPOTHESIS TESTING**

#### **Hypothesis 1: Higher Academic Success Efficiency (ASE) is associated with better student performance outcomes.**



```python
print("""Null Hypothesis (H0): There is no difference in mean ASE between successful students (Pass/Distinction) and at-risk students (Fail/Withdrawn)."
Alternative Hypothesis (H1): Successful students have significantly higher mean ASE than at-risk students.
Significance Level: alpha = 0.05 (5%)""")

# Statistical Analysis
successful = student_complete[student_complete['final_result'].isin(['Pass', 'Distinction'])]['ASE'].dropna()
at_risk = student_complete[student_complete['final_result'].isin(['Fail', 'Withdrawn'])]['ASE'].dropna()
difference = successful.mean() - at_risk.mean()
pct_diff = (difference / at_risk.mean()) * 100

print("\nDescriptive Statistics:")
print(f"Successful students (n={len(successful):})")
print(f"Mean ASE = {successful.mean():.2f}\n")

print(f"At-risk students (n={len(at_risk):})")
print(f"Mean ASE = {at_risk.mean():.2f}\n")

print(f"Absolute Difference = {difference:.2f}")
print(f"Percentage difference: {pct_diff:.2f}% higher for successful students")

# statistical test
print("\nStatistical Test:")
t_statistic, p_value = stats.ttest_ind(successful, at_risk)

print(f"T-Statistic: {t_statistic:.2f}")
print(f"P-Value: {p_value:.6f}")

# calculating effect
std_success = successful.std()
std_at_risk = at_risk.std()
n1 = len(successful)
n2 = len(at_risk)

pooled_std = np.sqrt(((n1 - 1) * std_success ** 2 + (n2 - 1) * std_at_risk ** 2) / (n1 + n2 - 2))
cohens_d = difference / pooled_std

if abs(cohens_d) < 0.2:
  effect = "Negligible effect"
elif abs(cohens_d)  < 0.5:
  effect = "Small effect"
elif abs(cohens_d) < 0.8:
  effect = "Medium effect"
else:
  effect = "Large effect"

# Hypothesis Decision
alpha = 0.05
print("\nDecision:")
if p_value < alpha:
  if effect=="Negligible effect":
    print(f"""P-value ({p_value:.6f}) < Alpha ({alpha})
Decision:  Reject the null hypothesis (H0)
Conclusion: H1 is statistically supported - however this is a non finding.
Cohen's d = {cohens_d:.2f} ({effect}) - the difference between successful and at-risk students is too small to carry any practical meaning""")
  else:
    print(f"""P-value ({p_value:.6f}) < Alpha ({alpha})
Decision: Reject the null hypothesis (H0)
Conclusion: Alternative Hypothesis (H1) is supported.""")
else:
  print(f"""P-value ({p_value:.6f}) >= Alpha ({alpha})
Decision: Fail to reject the null hypothesis (H0)
Conclusion: Alternative Hypothesis (H1) is rejected.""")

print(f"\nCohens_d: {cohens_d:.2f}")
print(f"Effect Size Interpretation: {effect}")

# visualization
fig, ax1 = plt.subplots(figsize=(8,6))
bp = ax1.boxplot([successful, at_risk],
                 tick_labels=['Successful\n(Pass + Distinction)', 'At-Risk\n(Fail + Withdrawn)'],
                 patch_artist=True)
for median in bp['medians']:
  median.set(color='black', linewidth=1.5)
bp['boxes'][0].set_facecolor('#27ae60')
bp['boxes'][1].set_facecolor('#e74c3c')
bp['boxes'][0].set_alpha(0.7)
bp['boxes'][1].set_alpha(0.7)

ax1.scatter([1, 2], [successful.mean(), at_risk.mean()], color='black', s=100, zorder=5, marker='D')
ax1.text(1.15, successful.mean(), f'Mean: {successful.mean():.2f}', fontsize=10, fontweight='bold', color='#27ae60')
ax1.text(2.15, at_risk.mean(), f'Mean: {at_risk.mean():.2f}', fontsize=10, fontweight='bold', color='#e74c3c')

# add significance annotation
sig_text = f'T-stat = {t_statistic:.2f}\np = {p_value:.6f}\nCohens_d = {cohens_d:.2f}\n{effect}'
ax1.annotate(sig_text, xy=(0.98, 0.98), xycoords='axes fraction', fontsize=10, fontweight='bold', va='top', ha='right', bbox=dict(boxstyle='round', facecolor='lightyellow', alpha=0.9, edgecolor='gray'))

ax1.set_title('H1: ASE by Performance\n(Does ASE Differentiate Success?)', fontsize=12, fontweight='bold')
ax1.set_ylabel('ASE')
ax1.set_xlabel('Performance')

plt.tight_layout()
plt.show()
```

#### **1. Statistical Findings**
* Successful students show higher Academic Success Efficiency (ASE), with a **mean ASE** of **(2.02)**, compared to **1.41 for at-risk students.**

* The **t-test result (t=54.01, p<0.001)** indicates that the difference in mean ASE between the two groups is statistically significant.

* The effect size is medium (Cohen'd ~ 0.61), suggesting that ASE meaningfully differentiates successful and at-risk students.


#### **2. Conclusion**
The **null hypothesis is rejected,** indicating that **ASE differs significantly between successful and at-risk students,** although the magnitude of the difference is small.


#### **3. Business Interpretation**
* **Academic Success Efficiency can be a useful performance indicator.** Students with higher ASE are more likely to achieve successful outcomes such as Pass or Distinction.

* **Monitoring ASE can help identify t-risk students early.** Students with lower ASE values may require targeted academic support to improve learning outcomes.

#### **Hypothesis 4: Strategic Assessment Preparation as a predictor of Academic Success.**

```python
print("""Null Hypothesis (H0): There is no difference in assessment preparartion ratio between successful and at-risk students.
Alternative Hypothesis (H1): There is significant difference in assesmnet preparation ration between successful and at-risk students.
Significance Level alpha = 0.05 (5%)""")

# Statistical Analysis
successful = student_complete[student_complete['final_result'].isin(['Pass', 'Distinction'])]['assessment_preparation_ratio'].dropna()
at_risk = student_complete[student_complete['final_result'].isin(['Fail', 'Withdrawn'])]['assessment_preparation_ratio'].dropna()
difference = successful.mean() - at_risk.mean()
pct_diff = (difference / at_risk.mean()) * 100 if at_risk.mean()!= 0 else 0

print("\nDescriptive Statistics")
print(f"Successful students (n={len(successful):})")
print(f"Mean Assessment Preparation Ratio = {successful.mean():.2f}\n")

print(f"At-risk students (n={len(at_risk):})")
print(f"Mean Assessment Preparation Ratio = {at_risk.mean():.2f}\n")

print(f"Absolute Difference = {difference:.2f}")
print(f"Percentage difference: {abs(pct_diff):.2f}% {'higher' if difference > 0 else 'lower'} for successful students")

# Statistical Test
print("\nStatistical Test:")
t_statistic, p_value = stats.ttest_ind(successful, at_risk)

print(f"T-Statistic: {t_statistic:.2f}")
print(f"P-Value: {p_value:.6f}")

# calculating effect size
std_successful = successful.std()
std_at_risk = at_risk.std()
n1 = len(successful)
n2 = len(at_risk)

pooled_std = np.sqrt(((n1 - 1) * std_successful ** 2 + (n2 - 1) * std_at_risk ** 2) / (n1 + n2 - 2))
cohens_d = difference / pooled_std

if abs(cohens_d) < 0.2:
  effect = "Negligible effect"
elif abs(cohens_d) < 0.5:
  effect = "Small effect"
elif abs(cohens_d) < 0.8:
  effect = "Medium effect"
else:
  effect = "Large effect"

# Hypothesis Decision
alpha = 0.05
print("\nDecision:")
if p_value < alpha:
  if effect=="Negligible effect":
    print(f"""P-value ({p_value:.6f}) < Alpha ({alpha})
Decision:  Reject the null hypothesis (H0)
Conclusion: H1 is statistically supported - however this is a non finding.
Cohen's d = {cohens_d:.2f} ({effect}) - the difference between successful and at-risk students is too small to carry any practical meaning""")
  else:
    print(f"""P-value ({p_value:.6f}) < Alpha ({alpha})
Decision: Reject the null hypothesis (H0)
Conclusion: Alternative Hypothesis (H1) is supported.""")
else:
  print(f"""P-value ({p_value:.6f}) >= Alpha ({alpha})
Decision: Fail to reject the null hypothesis (H0)
Conclusion: Alternative Hypothesis (H1) is rejected.""")

print(f"\nCohens_d: {cohens_d:.2f}")
print(f"Effect Size Interpretation: {effect}")

# visualization
fig, ax1 = plt.subplots(figsize=(8,6))
bp = ax1.boxplot([successful, at_risk], tick_labels=['Successful\n(Pass + Distinction)', 'At-Risk\n(Fail + Withdrawn)'], patch_artist=True)
for median in bp['medians']:
  median.set(color='black', linewidth=1.5)
bp['boxes'][0].set_facecolor('#1f77b4')
bp['boxes'][1].set_facecolor('#ff7f0e')
bp['boxes'][0].set_alpha(0.7)
bp['boxes'][1].set_alpha(0.7)

ax1.scatter([1, 2], [successful.mean(), at_risk.mean()], color='black', s=100, zorder=5, marker='D')
ax1.text(1.15, successful.mean(), f'Mean: {successful.mean():.2f}', fontsize=10, fontweight='bold', color='#1f77b4')
ax1.text(2.15, at_risk.mean(), f'Mean: {at_risk.mean():.2f}', fontsize=10, fontweight='bold', color='#ff7f0e')

# add significance annotation
sig_text = f'T-stat = {t_statistic:.2f}\np = {p_value:.6f}\nCohens_d = {cohens_d:.2f}\n{effect}'
ax1.annotate(sig_text, xy=(0.98, 0.98), xycoords='axes fraction', fontsize=10, fontweight='bold', va='top', ha='right', bbox=dict(boxstyle='round', facecolor='lightyellow', alpha=0.8, edgecolor='gray'))

ax1.set_title('H4: Assessment Preparation ratio by Outcome\n(Does Strategic Assessment Preparation Predict Success?)', fontsize=12, fontweight='bold')
ax1.set_ylabel('Assessment Preparation Ratio')
ax1.set_xlabel('Student Outcome')

plt.tight_layout()
plt.show()
```

#### **1. Statistical Findings**
* At-risk students show a **higher mean assessment preparation ratio (0.33**) compared to **successful students (0.18).**

* The **t-test result (t=-56.81, p<0.05)** indicates a significantly significant difference between the two groups.

* The **effect size (Cohens'd ~-0.68)** suggests a moderate practical difference in preparation patterns.


#### **2. Conclusion**
The **null hypothesis is rejected,** indicating that **assessment preparation ratios differ significantly between successful and at-risk students.**
